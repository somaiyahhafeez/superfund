# adding zip level population data

```python
zip_data=pd.read_csv("uszips.csv")
zip_data
```

```python
columns=["lat",	"lng",	"city",	"state_id",	"state_name",	"zcta",	"parent_zcta", "density",	"county_fips",	"county_name",	"county_weights",	"county_names_all",	"county_fips_all",	"imprecise",	"military",	"timezone"]
zip_data = zip_data.drop(columns=columns)
zip_data
```

```python
df = df.dropna(subset=["Zip"])
df['Zip'] = df['Zip'].astype(float).astype(int).astype(str).str.zfill(5)

zip_data['zip'] = zip_data['zip'].astype(str).str.zfill(5)
```

```python
merged = pd.merge(
    df,
    zip_data, 
    left_on='Zip',
    right_on='zip',
    how='left'
)
```

```python
merged
```

```python
merged['Site Name'] = merged['Site Name'].str.title()
```

```python
state_counts = merged.groupby("State Name")["Site Name"].count().reset_index(name="Site Count")
state_counts
```

```python
merged['Year'] = pd.to_datetime(merged['NPL Status'], format='%m/%d/%y').dt.year

merged['City'] = merged['City'].str.title()
merged
```

```python
columns=["NPL Status", "Year Added to NPL"]
merged = merged.drop(columns=columns)
merged
```

```python
merged.to_csv("superfund_map.csv")
```
